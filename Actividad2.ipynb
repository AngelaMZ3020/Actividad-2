```python
# Cargar a la memoria de trabajo las bibliotecas: Pandas (manejo de datos, 
# Numpy (numerical python) y la biblioteca de gráficas Matplotlib
# Se asignan nombres cortos.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#
# Usar "Shift+Enter" para procesar la información de la celda
#
```

```python
# Descarga los datos de una estación del Servicio Meteorológico Nacional
# http://smn1.conagua.gob.mx/emas/
# Lee un archivo de texto con la función Pandas "read_csv", con elementos separados por mas de 
# un espacio, brincándose 4 renglones del inicio (encabezados)
df0 = pd.read_csv('Kino.txt', skiprows=4, sep='\s+')
# "Shift + Enter"
```

```python
# Lee los primeros 5 renglones del archivo
df0.head()
# "Shift+Enter"
```

```python
# Dar estructura de datos (DataFrame)
df = pd.DataFrame(df0)
```

```python
# Ver los tipos de datos que Pandas ha reconocido al leer
df.dtypes
```

```python
# Combinar las columnas "DD/MM/AAAA" con "HH:MM" y convertirla a variable de tiempo
# Se crea una nueva columna "Fecha" al final con formato de tiempo.
# Eliminamos las dos primeras columnas que ya no necesitaremos
df['FECHA'] = pd.to_datetime(df.apply(lambda x: x['DD/MM/AAAA'] + ' ' + x['HH:MM'], 1), dayfirst=True)
df = df.drop(['DD/MM/AAAA', 'HH:MM'], 1)
```

```python
df.head()
```

```python
# Realiza un análisis exploratorio de datos
df.describe()
```

```python
# Selecciona los renglones con Temperatura > 24ºC y < 25ºC
df_tmp = df[df.TEMP > 24] 
df_select = df_tmp[df_tmp.TEMP < 25]
df_select
```

```python
# Calcula el promedio de las columnas, excepto en la FECHA (que no tendría sentido)
df.mean()
```

```python
# Calcula el promedio de las Temperaturas
df.TEMP.mean()
```

```python
# Gráfica de la rapidez de los vientos (m/s) 
plt.figure(); df.VELS.plot(); plt.legend(loc='best')
plt.title("Variación de la Rapidez de los Vientos")
plt.ylabel("Rapidez (m/s)")
plt.grid(True)
plt.show()
```

```python
# Gráfica de Temperatura y Humedad Relativa
df1 = df[['TEMP','HR']]
plt.figure(); df1.plot(); plt.legend(loc='best')
plt.title("Variación de la Temperatura y la Humedad Relativa")
plt.ylabel("Temp ºC /(%) HR")
plt.grid(True)
plt.show()
```

```python
plt.plot_date(x=df.FECHA, y=df.TEMP, fmt="b-")
plt.title("Variación de la Temperatura")
plt.ylabel("Temp ºC")
plt.grid(True)
plt.show()
```

```python
#Gráfica que muestre la rapidez de los vientos (VELS) y la rapidez de las ráfagas (VELR) como funciones del tiempo.
df2 = df[['VELS','VELR']]
plt.figure(); df2.plot(); plt.legend(loc='best')
plt.title("Rapidez de los vientos (VELS) y rapidez de las ráfagas (VELR)")
plt.ylabel("VELS (m/s) / VELR (m/s)")
plt.grid(True)
plt.show()
```

```python
#Gráfica de la dirección de los vientos (DIRS) como función del tiempo.
plt.figure(); df.DIRS.plot();plt.legend(loc='best')
plt.title("Dirección de los Vientos")
plt.ylabel("Dirección")
plt.grid(True)
plt.show()
```

```python
#Se puede observar que los vientos dominantes están en la dirección alrededor de los 250.
#Tal como lo indica la media, la cual es:
df.DIRS.mean()
```

```python
#Gráfica del comportamiento de la Radiación Solar como función del tiempo.
plt.figure();df.RADSOL.plot();plt.legend(loc='best')
plt.title("Radiación Solar con respecto al tiempo")
plt.ylabel("Radiación Solar")
plt.grid(True)
plt.show()
```

```python
#En Kino, tal como la gráfica lo muestra, puede observarse que los niveles de radiación solar aumentan y disminuyen
#cada cierto tiempo, casi de manera periodica, con lo que se puede concluir que en el transcurso de de esos 5 días
#el nivel de radiación es mayor en el día que en la noche.
```

```python
#Lapso de la temperatura diaria
df.TEMP.max() - df.TEMP.min()
```

```python
#Relación entre la temperatura y la humedad relativa
df1 = df[['TEMP','HR']]
plt.figure(); df1.plot(); plt.legend(loc='best')
plt.title("Variación de la Temperatura y la Humedad Relativa")
plt.ylabel("Temp ºC /(%) HR")
plt.grid(True)
plt.show()
```

```python
#A lo que se puede concluir en relación a la tabla de temperatura y humedad relativa es que, en la mayoría de las horas
#de alrededor de 5 días, la temperatura era mayor que la humedad relativa, y solo en pocas horas la humedad relativa
#superaba a la temperatura en el momento.
```
